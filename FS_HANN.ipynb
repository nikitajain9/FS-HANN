```python
#    Importing all the libraries
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif, chi2
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, f1_score, confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

#    Loading and Setting up the data
print("Loading Data...")
df = pd.read_csv('heart_disease_health_indicators_BRFSS2015.csv')


X = df.drop(columns=["HeartDiseaseorAttack"])
y = df["HeartDiseaseorAttack"]

#    Spilting the data into train, test data
X_train_raw, X_test_raw, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

#    FS-HANN Model Builder
def build_fs_hann(input_dim):

    model = Sequential([
        Dense(128, activation='relu', input_dim=input_dim),
        BatchNormalization(),
        Dropout(0.3),

        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),

        Dense(32, activation='relu'),
        BatchNormalization(),

        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=Adam(learning_rate=1e-3), # Started slightly higher, let ReduceLROnPlateau handle decay
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.AUC(name='auc')]
    )
    return model

#    Experimenting
def run_experiment(k_features, imbalance_strat, fs_method='mutual_info'):
    print(f"\n--- Running: K={k_features} | Strat={imbalance_strat} | FS={fs_method} ---")

    #    Feature Selection
    #    Note: MinMaxScaler is safer for Neural Networks and required for Chi2
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train_raw)
    X_test_scaled = scaler.transform(X_test_raw)

    start_fs = time.time()
    if fs_method == 'mutual_info':
        # Using a subset for mutual_info speed if dataset > 50k rows
        selector = SelectKBest(score_func=mutual_info_classif, k=k_features)
    else:
        selector = SelectKBest(score_func=chi2, k=k_features)

    X_train_sel = selector.fit_transform(X_train_scaled, y_train)
    X_test_sel = selector.transform(X_test_scaled)
    fs_time = time.time() - start_fs

    #    Get selected feature names
    cols = X.columns[selector.get_support(indices=True)]
    print(f"Selected Features: {list(cols)}")

    #    Imbalance Handling
    class_weights = None
    X_train_final = X_train_sel
    y_train_final = y_train

    if imbalance_strat == 'smote':
        print("Applying SMOTE...")
        sm = SMOTE(random_state=42)
        X_train_final, y_train_final = sm.fit_resample(X_train_sel, y_train)
    elif imbalance_strat == 'class_weight':
        classes = np.unique(y_train)
        cw = compute_class_weight('balanced', classes=classes, y=y_train)
        class_weights = {classes[i]: cw[i] for i in range(len(classes))}
        print(f"Class Weights calculated: {class_weights}")

    #    Model Training
    model = build_fs_hann(input_dim=k_features)

    rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2, min_lr=1e-6)
    es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    start_train = time.time()
    history = model.fit(
        X_train_final, y_train_final,
        validation_split=0.2,
        epochs=30, # Reduced epochs for comparative speed (early stopping will catch it)
        batch_size=64,
        class_weight=class_weights,
        callbacks=[rlr, es],
        verbose=0 # Silent mode to keep output clean
    )
    train_time = time.time() - start_train

    #    Evaluation
    y_prob = model.predict(X_test_sel).ravel()
    y_pred = (y_prob >= 0.5).astype(int)

    results = {
        'K': k_features,
        'Strategy': imbalance_strat,
        'FS_Method': fs_method,
        'Accuracy': accuracy_score(y_test, y_pred),
        'AUC': roc_auc_score(y_test, y_prob),
        'Recall': recall_score(y_test, y_pred),
        'Precision': classification_report(y_test, y_pred, output_dict=True)['1.0']['precision'],
        'F1': f1_score(y_test, y_pred),
        'Train_Time': train_time,
        'FS_Time': fs_time
    }
    print(f"Result -> AUC: {results['AUC']:.4f} | Recall: {results['Recall']:.4f} | Time: {results['Train_Time']:.2f}s")
    return results

#    Execution Loop
results_list = []

# Scenarios to test for the Comparative Analysis
scenarios = [
    (10, 'class_weight', 'chi2'), # Faster baseline
    (15, 'class_weight', 'chi2'),
    (21, 'class_weight', 'chi2'), # Full features
    (15, 'smote', 'chi2'),        # Comparison with SMOTE
    # (15, 'class_weight', 'mutual_info') # commented because slow on large data
]

for k, strat, fs in scenarios:
    res = run_experiment(k, strat, fs)
    results_list.append(res)

# Result
res_df = pd.DataFrame(results_list)
print("\nFinal Comparative Table:")
print(res_df)


plt.figure(figsize=(10, 6))
sns.barplot(data=res_df, x='K', y='AUC', hue='Strategy')
plt.title("Gap Analysis: Impact of Feature Selection & Imbalance Strategy")
plt.ylim(0.7, 0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


try:
    res_df
except NameError:
    raise RuntimeError("res_df not found in workspace. Run your experiments loop first to produce res_df.")

# 1) Clean/format numeric columns for plotting
plot_df = res_df.copy()
# ensure numeric types
for c in ['AUC','Recall','F1','Train_Time','FS_Time','Accuracy','Precision']:
    if c in plot_df.columns:
        plot_df[c] = pd.to_numeric(plot_df[c], errors='coerce')

# 2) Comparative bar plots for key metrics
metrics = ['AUC','Recall','F1','Train_Time','FS_Time']
plt.rcParams.update({'figure.dpi': 150, 'font.size': 11})

for metric in metrics:
    if metric not in plot_df.columns:
        continue
    plt.figure(figsize=(7,4.2))
    sns.barplot(data=plot_df, x='K', y=metric, hue='Strategy', ci=None)
    plt.title(f'Comparative: {metric} by K and Strategy')
    if metric in ['AUC','Recall','F1','Accuracy','Precision']:
        plt.ylim(0,1.0)
    plt.tight_layout()
    fname = f'plot_{metric.lower()}.png'
    plt.savefig(fname, bbox_inches='tight', dpi=300)
    print(f"Saved: {fname}")
    plt.show()

# 3) Results table -> CSV + PNG
res_csv = 'results_summary.csv'
plot_df.to_csv(res_csv, index=False)
print(f"Saved CSV: {res_csv}")

# create a PNG of the table (matplotlib table)
def save_table_image(df, filename='results_table.png', fontsize=10, col_width=2.8):
    fig, ax = plt.subplots(figsize=(min(12, col_width*len(df.columns)), 0.6*len(df)+1))
    ax.axis('off')
    tbl = ax.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='center')
    tbl.auto_set_font_size(False)
    tbl.set_fontsize(fontsize)
    tbl.scale(1, 1.2)
    plt.title('Comparative Results Table', pad=12)
    plt.savefig(filename, bbox_inches='tight', dpi=300)
    plt.show()
    print(f"Saved table image: {filename}")

save_table_image(plot_df, filename='results_table.png', fontsize=10)

# 4) Heatmap of metrics (normalized)
metric_cols = [c for c in ['AUC','Recall','F1','Train_Time'] if c in plot_df.columns]
if metric_cols:
    # normalize metrics so heatmap is comparable (Train_Time invert so lower=better)
    heat_df = plot_df.set_index(['K','Strategy'])[metric_cols].copy()
    # invert Train_Time so higher is better visually
    if 'Train_Time' in heat_df.columns:
        heat_df['Train_Time_inv'] = -heat_df['Train_Time']
        heat_df = heat_df.drop(columns=['Train_Time']).rename(columns={'Train_Time_inv':'Train_Time (inv)'})
    # z-score normalize per column
    heat_norm = (heat_df - heat_df.mean())/heat_df.std()
    plt.figure(figsize=(8, max(3, heat_norm.shape[0]*0.25)))
    sns.heatmap(heat_norm, annot=True, fmt='.2f', cmap='vlag', cbar_kws={'label':'z-score'})
    plt.title('Heatmap (z-scored) of Metrics by (K,Strategy)')
    plt.tight_layout()
    plt.savefig('metrics_heatmap.png', dpi=300, bbox_inches='tight')
    print("Saved: metrics_heatmap.png")
    plt.show()

# 5) Pairwise scatter plots to show trade-offs (AUC vs Train_Time, Recall vs Train_Time)
pairs = [('AUC','Train_Time'), ('Recall','Train_Time'), ('F1','Train_Time'), ('Recall','AUC')]
for x,y in pairs:
    if x in plot_df.columns and y in plot_df.columns:
        plt.figure(figsize=(6,5))
        sns.scatterplot(data=plot_df, x=x, y=y, hue='Strategy', style='K', s=130)
        for i, row in plot_df.iterrows():
            plt.text(row[x], row[y], f"K={row['K']}", fontsize=9, horizontalalignment='left', verticalalignment='bottom')
        plt.title(f'{y} vs {x}')
        plt.tight_layout()
        fname=f'scatter_{x.lower()}_{y.lower()}.png'
        plt.savefig(fname, dpi=300, bbox_inches='tight')
        print(f"Saved: {fname}")
        plt.show()
```
